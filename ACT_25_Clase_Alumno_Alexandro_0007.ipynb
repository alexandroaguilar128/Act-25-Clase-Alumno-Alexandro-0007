### Clase Alumno
### Descripcion:
Breve explicación 📝
La Clase (Alumno) 🏫: Funciona como una plantilla o plano maestro. Aquí definimos que cualquier alumno que registremos debe tener obligatoriamente tres características: nombre, matrícula y promedio.

El Constructor (__init__) 🏗️: Es el punto de partida. Se asegura de que, en cuanto creamos a un alumno, sus "contenedores" de información estén listos para recibir datos, aunque al inicio estén vacíos.

Los Métodos (Acciones) ⚙️:

leerdatos(): Es la función de entrada ⌨️. Permite que el usuario interactúe con el programa para asignar valores específicos a los atributos del objeto.

mostrardatos(): Es la función de salida 🖥️. Toma la información guardada y la muestra de forma ordenada en la consola.

Los Objetos (alumno1 y alumno2) 👥: Son las instancias reales. Aunque ambos usan el mismo código, son totalmente independientes; lo que escribas para el alumno1 no afectará la información almacenada en el alumno2.

El uso de self 📍: Es la forma en que Python identifica que los datos pertenecen específicamente al alumno que estamos consultando en ese momento y no a otro.

![Info](https://raw.githubusercontent.com/alexandroaguilar128/Act-25-Clase-Alumno-Alexandro-0007/refs/heads/main/infoclassxxx.png)

```python
class Alumno:
    def __init__(self):
        self.nombre = ""
        self.matricula = ""
        self.promedio = 0.0

    def leerdatos(self):
        self.nombre = input("Ingrese el nombre: ")
        self.matricula = input("Ingrese la matrícula: ")
        self.promedio = float(input("Ingrese el promedio: "))

    def mostrardatos(self):
        # Imprime la fila con formato alineado
        print(f"{self.nombre:<15} | {self.matricula:<12} | {self.promedio:<8}")

# Creación de objetos
alumno1 = Alumno()
alumno2 = Alumno()

# Entrada de datos
print("--- Registro de Alumno 1 ---")
alumno1.leerdatos()
print("\n--- Registro de Alumno 2 ---")
alumno2.leerdatos()

# Salida en formato de tabla
print("\n" + "="*43)
print(f"{'Nombre':<15} | {'Matrícula':<12} | {'Promedio':<8}")
print("-" * 43)
alumno1.mostrardatos()
alumno2.mostrardatos()
print("="*43)
print(" Programa realizado por Alexandro Aguilar NL = 1 ")
```
